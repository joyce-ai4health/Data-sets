```python
# DATA CLEANING PHASE
### Step 1: Calculate the Median BMI

# - Calculate the median (middle value) of the BMI column
# - We do this before filling missing values because we need a replacement value.
# - The median is preferred over the mean because BMI data may contain extreme values (outliers) that could distort the average.

### NOTE
# In medical datasets like stroke prediction:

# - Some patients may have very high BMI (obesity cases)
# - Some may have very low BMI (malnutrition or illness cases)
# - These extreme values are medically valid but can skew averages

# So using the mean: may "inflate" or "deflate" the typical BMI of patients.

# But using the  median: gives more realistic “central patient”
```

```python
bmi_median = df["bmi"].median()
print("Median BMI: " , bmi_median)
```

```python
## Step 2: Fill missing bmi values
# - Replace all missing BMI values (NaN) with the median BMI value(28.1)
# - This ensures every patient record has a BMI value.
# - We assign the result back to the BMI column so the changes are saved.
df["bmi"] = df["bmi"].fillna(bmi_median)
```

```python
## Step 3: Verify missing values has been removed
# - Count missing values again after filling BMI.
# - This confirms that the cleaning step worked correctly.
df.isnull().sum()
```

```python
## BMI Missing Value Treatment

# The BMI variable contained 201 missing values, representing approximately 3.93% of the dataset.

# Rather than removing these patient records, median imputation was used to preserve the dataset size and minimize information loss.

# The median was selected because BMI distributions can contain extreme values (outliers), making it a more robust measure of central tendency than the mean.

# After imputation, all missing values were successfully removed, resulting in a complete dataset for subsequent exploratory data analysis and predictive modelling.
```

```python
### Step 6: Assess Stroke Class Balance
# Count the number of observations
# in each stroke category.

stroke_counts = df["stroke"].value_counts()

print(stroke_counts)
```

```python
# Calculate percentage distribution
# of each stroke category.

# Reason:
# Healthcare datasets frequently contain
# imbalanced target variables.

# In this dataset,
# stroke cases are expected to be much less frequent
# than non-stroke cases.

# Understanding class balance helps us decide
# whether special techniques such as SMOTE
# will be required during model development.

# Class imbalance can cause a model
# to become biased toward predicting
# the majority class.

stroke_percentages = (
    df["stroke"]
    .value_counts(normalize=True)
    * 100
)

print(stroke_percentages)
```

```python
# STROKE CLASS DISTRIBUTION

# The target variable is highly imbalanced.

# The majority of patients did not experience a stroke,
# while only a small percentage experienced a stroke event.

# This is common in healthcare datasets
# because severe medical outcomes occur less frequently
# than normal outcomes.

# To improve predictive performance,
# class balancing techniques such as SMOTE
# may be considered during feature engineering.
```

```python

### STEP 7: Remove ID Column


# The id column is a unique patient identifier.

# It does not contain any clinical,
# demographic, or behavioural information.

# Therefore, it provides no predictive value
# for stroke risk analysis.

# Keeping identifier columns may introduce noise
# into machine learning models.

# For this reason, the id column is removed.

df = df.drop(columns=["id"])
```
