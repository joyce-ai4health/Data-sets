```python
## STEP 1: Pair Plot (Scatterplot Matrix)

# Create a pair plot (scaerplot matrix) to visualize the relationships between the dierent features
# (sepal length, sepal width, petal length, and petal width).
# Reason: this is the first required visualization — it shows scatterplots for
# every pair of features, help in seeing which features separate species well
# hue='species': colors each point by species so we can visually spot clustering
# diag_kind='hist': shows a histogram on the diagonal instead of a density curve

pair_plot = sns.pairplot(iris, hue='species', diag_kind='hist')

# Add an overall title above the grid of plots
# Reason: makes the figure self-explanatory when viewed on its own in the report
pair_plot.fig.suptitle("Pair Plot of Iris Features by Species", y=1.02)

# Render the plot
# Reason: without this, the plot may not display depending on the environment
plt.show()
```

```python
## STEP 2: Boxplots (Distribution per Species)

#  Use boxplots to visualise the distribution of each feature across the three species.

# Create a figure with 4 subplots (one for each numeric feature)
# Reason: lets us compare the distribution of each feature across species
# side-by-side in a single image
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# List of the four numeric feature column names
# Reason: we'll loop through these to avoid repeating code four times
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# axes.flatten(): converts the 2x2 grid of subplot axes into a simple list
# Reason: makes it easy to loop through and assign one feature per subplot

for ax, feature in zip(axes.flatten(), features):

    # Draw a boxplot of the current feature, grouped by species
    # Reason: boxplots show median, quartiles, and outliers per species,
    # which helps identify which features best distinguish the species

    sns.boxplot(data=iris, x='species', y=feature, ax=ax)

    # Set a title for this specific subplot
    # Reason: clarifies which feature this particular boxplot represents
    ax.set_title(f'Boxplot of {feature.replace("_", " ").title()} by Species')

# plt.tight_layout(): automatically adjusts spacing so subplot titles/labels
# don't overlap each other
# Reason: improves readability of the final figure
plt.tight_layout()

# Render the plots
plt.show()
```

```python
## STEP 3: Histograms (Numerical Variables)

# Generate histograms for the numerical variables (sepal length, sepal width, etc.).
# Create a figure with 4 subplots, one histogram per numeric feature
# Reason: required by the assessment — histograms show the frequency
# distribution (shape, spread, skewness) of each measurement
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Loop through each feature and its corresponding subplot axis
for ax, feature in zip(axes.flatten(), features):
    # Draw a histogram with a kernel density estimate (KDE) line overlaid
    # Reason: the KDE line makes it easier to see the overall distribution
    # shape on top of the raw bar counts
    sns.histplot(data=iris, x=feature, hue='species', kde=True, ax=ax, element='step')

    # Set a title for this subplot
    ax.set_title(f'Distribution of {feature.replace("_", " ").title()}')

# Adjust spacing so labels/titles don't overlap
plt.tight_layout()

# Render the plots
plt.show()
```

```python
## STEP 4: Additional Plot — Correlation Heatmap

# Implement at least one other plot of your choice that helps visualise the data.

# Select only the numeric columns for correlation calculation
# Reason: correlation only makes sense between numeric variables, so we
# drop the categorical 'species' column first
numeric_data = iris.drop(columns='species')

# Calculate the pairwise correlation coefficients between all numeric features
# Reason: tells us how strongly each feature moves together with another
# (e.g., petal length and petal width are usually highly correlated)
correlation_matrix = numeric_data.corr()

# Set the figure size for the heatmap
plt.figure(figsize=(8, 6))

# Draw the heatmap of the correlation matrix
# annot=True: prints the actual correlation numbers inside each cell
# cmap='coolwarm': uses a red-blue color scale so high/low values are obvious
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add a title
plt.title("Correlation Heatmap of Iris Features")

# Render the plot
plt.show()
```

# **4. Simple Classification Model:**

● Split the dataset into training and test sets (e.g., 70%
training and 30% testing).

● Implement a k-Nearest Neighbors (k-NN) classifier to
predict the species of the iris flowers.

● Evaluate the accuracy of your model on the test set and
display the confusion matrix.

● Provide a brief explanation of the k-NN algorithm and
interpret the model’s results

```python
## STEP 1: Prepare Data for Classification
# Split the dataset into training and test sets (e.g., 70% training and 30% testing).

# Separate the features (X) from the target label (y)
# Reason: machine learning models need inputs (X) and the answer to predict (y)
# kept separately
X = iris.drop(columns='species')   # X = the 4 numeric measurements
y = iris['species']                # y = the species label we want to predict

# Split the data into training (70%) and testing (30%) sets
# Reason: required by the assessment — training data teaches the model,
# test data evaluates it on examples it has never seen, giving an honest
# measure of performance
# random_state=42: fixes the randomness so the split is reproducible every run
# stratify=y: ensures each species is proportionally represented in both
# the training and test sets (avoids an unlucky imbalanced split)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Print the resulting sizes to confirm the split worked as expected
print("Training set size:", X_train.shape[0])
print("Testing set size:", X_test.shape[0])
```

```python
## STEP 2:Build and Train the k-NN Classifier
#  Implement a k-Nearest Neighbors (k-NN) classifier to predict the species of the iris flowers.

# Create a k-Nearest Neighbors classifier
# Reason: required by the assessment — k-NN classifies a new point by looking
# at the 'k' closest labeled points in the training data and taking a majority vote
# n_neighbors=5: a common default choice for k; means each prediction looks at
# the 5 nearest training samples
knn = KNeighborsClassifier(n_neighbors=5)

# Train (fit) the model using the training data
# Reason: this is where the model "learns" by storing/indexing the training
# samples so it can later measure distances to them
knn.fit(X_train, y_train)

# Use the trained model to predict species for the unseen test data
# Reason: this lets us check how well the model generalizes to new data
y_pred = knn.predict(X_test)

# Display the actual vs predicted labels for a quick visual check
print("Actual labels:   ", list(y_test[:10]))
print("Predicted labels:", list(y_pred[:10]))
```

```python
## STEP 3: Evaluate the Model
# Evaluate the accuracy of your model on the test set and display the confusion matrix.

# Calculate the accuracy: proportion of correct predictions out of all predictions
# Reason: required by the assessment — gives a single overall performance score
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

# Generate the confusion matrix
# Reason: required by the assessment — shows exactly how many predictions were
# correct vs. which species got confused with which other species
cm = confusion_matrix(y_test, y_pred, labels=knn.classes_)

# Display the confusion matrix as a heatmap for easier interpretation
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=knn.classes_, yticklabels=knn.classes_)
plt.xlabel("Predicted Species")
plt.ylabel("Actual Species")
plt.title("Confusion Matrix - k-NN Classifier")
plt.show()

# Print a detailed classification report
# Reason: shows precision, recall, and f1-score per species — more detailed
# than accuracy alone, useful for discussing model performance per class
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

# **STEP 4: Explanation of k-NN (for your report's Methodology/Results section):**

Provide a brief explanation of the k-NN algorithm and interpret the model’s results

k-Nearest Neighbors is a simple, non-parametric algorithm that classifies a new data point by finding the 'k' most similar points (neighbors) in the training set, based on distance (typically Euclidean), and assigning the majority class among those neighbors.

It makes no assumptions about the underlying data distribution, which makes it easy to understand and apply, though it can be slower on very large datasets since it computes distances to many points at prediction time.
